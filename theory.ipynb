## Binary Search

```python
import numpy as np

def binary_search(target, nums):
    low, high = 0, len(nums) - 1
    count = 0

    while low <= high:
        mid = (low + high) // 2
        guess = nums[mid]
        count += 1

        if target == guess:
            return mid, count

        if guess < target:
            low = mid + 1
        else: 
            high = mid - 1
    
    return "Not in array"
        
nums = np.arange(45, 101)

print(binary_search(76, nums))
```

## Selection Sort

```python
def find_smallest(arr):
    smallest = arr[0]
    smallest_idx = 0

    for i in range(1, len(arr)):
        if arr[i] < smallest:
            smallest = arr[i]
            smallest_idx = i
    
    return smallest_idx

def selection_sort(arr):
    new_arr = []

    for i in range(len(arr)):
        smallest_idx = find_smallest(arr)
        new_arr.append(arr.pop(smallest_idx))
    
    return new_arr

arr = [14, 5, 6, 19, 2, 34, 65, 13, 27, 37]

print(selection_sort(arr))
```
