## Binary Search

```python
import numpy as np

def binary_search(target, nums):
    low, high = 0, len(nums) - 1
    count = 0

    while low <= high:
        mid = (low + high) // 2
        guess = nums[mid]
        count += 1

        if target == guess:
            return mid, count

        if guess < target:
            low = mid + 1
        else: 
            high = mid - 1
    
    return "Not in array"
        
nums = np.arange(45, 101)

print(binary_search(76, nums))
```

## Selection Sort

```python
def find_smallest(arr):
    smallest = arr[0]
    smallest_idx = 0

    for i in range(1, len(arr)):
        if arr[i] < smallest:
            smallest = arr[i]
            smallest_idx = i
    
    return smallest_idx

def selection_sort(arr):
    new_arr = []

    for i in range(len(arr)):
        smallest_idx = find_smallest(arr)
        new_arr.append(arr.pop(smallest_idx))
    
    return new_arr

arr = [14, 5, 6, 19, 2, 34, 65, 13, 27, 37]

print(selection_sort(arr))
```

## Recursion

```python
def countdown(i):
    print(i)

    if i <= 0:
        return None
    else:
        return countdown(i - 1)
    
countdown(10)
```

```python
def fact(x):

    if x == 1:
        return 1
    
    else:
        return x * fact(x - 1)
    
fact(3)
```

```python
def recur_sum(arr):
    if not arr:
        return 0
    
    if len(arr) == 1:
        return arr[0]
    
    first = arr.pop(0)

    return first + recur_sum(arr)

arr = [2, 4, 6, 18]

recur_sum(arr)
```

```python
def count_items(arr):
    if not arr:
        return 0

    count = 0
    
    arr.pop(0)
    count += 1
    
    return count + count_items(arr)
    
arr = [2, 4, 6, 18]

count_items(arr)
```

## Quick Sort

```python
def quicksort(arr):
    if len(arr) < 2:
        return arr
    
    pivot = arr[0]

    left_arr = []
    right_arr = []

    for i in arr[1:]:
        if i < pivot:
            left_arr.append(i)
        else:
            right_arr.append(i)
    
    print(left_arr, [pivot], right_arr)

    return quicksort(left_arr) + [pivot] + quicksort(right_arr)

arr = [3,5,2,1,4]

print(quicksort(arr))
```

```python
import random

def quicksort2(arr, low=0, high=None):
    if high is None:          
        high = len(arr) - 1

    if low < high:
        p = partition(arr, low, high)
        quicksort2(arr, low, p - 1)
        quicksort2(arr, p + 1, high)

    return arr                

def partition(arr, low, high):
    # pick a random pivot, swap it to the end so the rest assumes pivot == arr[high]
    pivot_index = random.randint(low, high)
    arr[pivot_index], arr[high] = arr[high], arr[pivot_index]
    pivot = arr[high]

    i = low - 1               # boundary of the "< pivot" region
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]   # pivot into its final slot
    return i + 1

arr = [3, 5, 2, 1, 4]
quicksort2(arr)
print(arr)
```
