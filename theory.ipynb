## Binary Search

```python
def binary_search(target, nums):
    low, high = 0, len(nums - 1)

    while low <= high:
        mid = (low + high) / 2
        guess = nums[mid]

        if guess == nums[mid]:
            return mid
        
        if guess < target:
            low = mid + 1
        else: 
            high = mid - 1
        
nums = []
```
